```python
# KMeans Clustering

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_wine
dataset = load_wine()
data = pd.DataFrame(dataset.data, columns = dataset.feature_names)
df = data.copy()

kmeans = KMeans(n_clusters = 3)
kmeans.fit(df)
pred = kmeans.predict(df)

df['target'] = dataset.target
df['pred'] = pred
df
```

```python
# PCA
df = data.copy()
pca = PCA(n_components = 2)
df = pca.fit_transform(df)
df = pd.DataFrame(df)
df.head()
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(df)
pred = kmeans.predict(df)
df['kmeans'] = pred
df
```

```python
centers = kmeans.cluster_centers_
```

```python
sns.scatterplot(data = df, x = 0, y = 1, hue = 'kmeans', palette = 'viridis')
plt.scatter(centers[:, 0], centers[:, 1], s = 200, c = 'r', marker = '*', label = 'center')
```

```python
df.columns = df.columns.astype(str)
inertia = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters = i, random_state = 2022)
    kmeans.fit(df)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 10), inertia, marker = 'o', linestyle = '--', color = 'b')    
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_circles, make_moons, make_blobs

# Set a random seed for reproducibility
np.random.seed(0)

# Generate synthetic data for the three different shapes

# 1. Concentric circles
circles_X, _ = make_circles(n_samples=300, factor=.5, noise=.05)

# 2. Overlapping crescent shapes (moons)
moons_X, _ = make_moons(n_samples=300, noise=.05)

# 3. Linear shapes (blobs)
blobs_X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.5, random_state=0)

# Plotting the datasets
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

# Concentric circles
axs[0].scatter(circles_X[:, 0], circles_X[:, 1])
axs[0].set_title('Concentric Circles')

# Overlapping crescent shapes
axs[1].scatter(moons_X[:, 0], moons_X[:, 1])
axs[1].set_title('Overlapping Crescent Shapes')

# Linear shapes
axs[2].scatter(blobs_X[:, 0], blobs_X[:, 1])
axs[2].set_title('Linear Shapes')

# Show the plot
plt.tight_layout()
plt.show()
```
