# Data preprosession with second hand car price dataset

## 1. Importing libraries and load datasets

```python
import numpy as np
import pandas as pd

brand = pd.read_csv('~/aiffel/data/brand.csv')
cars = pd.read_csv('~/aiffel/data/cars.csv')
```

```python
brand
```

```python
cars
```

```python
brand['title'].unique()
```

```python
cars['title'].unique()
```

## 2. Merge cars.csv and brand.csv based on the column 'title'

```python
# Extract and lower the first word of the title column
cars['title'] = cars['title'].str.lower().str.split().str[0]
```

```python
cars.head()
```

```python
# Merge two dataframes on the title column
cars_df = cars.merge(brand, on = 'title', how = 'left')
```

```python
cars_df
```

```python
cars_df.isna().sum()
```

```python
cars_df.info()
```

## 3. Change categorical data type to numeric type ('Engine', 'Emission Class')

```python
cars_df.describe()
```

```python
cars_df['Engine'] = cars_df['Engine'].str.replace('L', '')
```

```python
cars_df['Engine'] = cars_df['Engine'].astype(float)
```

```python
cars_df['Emission Class'] = cars_df['Emission Class'].str.replace('Euro', '')
```

```python
cars_df['Emission Class'] = cars_df['Emission Class'].astype(float)
```

```python
cars_df.describe()
```

```python
cars_df.info()
```

## 4. Handling Missing Values

```python
cars_df.isna().sum()
```

```python
cars_df['Engine'].fillna(cars_df['Engine'].mean(), inplace = True)
```

```python
cars_df['Emission Class'].fillna(cars_df['Emission Class'].mean(), inplace = True)
```

```python
cars_df['Doors'].fillna(cars_df['Doors'].mean(), inplace = True)
```

```python
cars_df['Seats'].fillna(cars_df['Seats'].mean(), inplace = True)
```

```python
cars_df['Previous Owners'].fillna(cars_df['Previous Owners'].mean(), inplace = True)
```

```python
cars_df['Service history'].fillna('Unknown', inplace = True)
```

```python
cars_df.isna().sum()
```

```python
cars_df.info()
```

## 7. Correlation Analysis

```python
cars_df.corr()
```

```python
# Import libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Understand correlation between features
colormap = plt.cm.coolwarm
plt.figure(figsize=(30, 30))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(cars_df.astype(float).corr(), linewidths=0.1, vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True, annot_kws={'size': 7}, fmt='.1f')
```

```python
# Calculate correlation between features
corr_matrix = cars_df.corr()
```

```python
# Treat NaN values in correlation matrix
corr_matrix.values[np.tril_indices_from(corr_matrix)] = np.nan
```

```python
# Sort correlation matrix by descending order
corr_matrix = corr_matrix.unstack().dropna().sort_values(key=np.abs, ascending=False).head(30)
```

```python
# Print correlation matrix in descending order
print(corr_matrix)
```
