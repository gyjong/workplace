```python
data = {
    'Product_ID': ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'P8', 'P9', 'P10'],
    'Category': ['Electronics', 'Electronics', 'Clothing', 'Clothing', 'Home', 'Home', 'Electronics', 'Electronics', 'Clothing', 'Clothing'],
    'Subcategory': ['Smartphones', 'Laptops', 'Dresses', 'Shirts', 'Furniture', 'Decor', 'Tablets', 'Smartphones', 'Shirts', 'Dresses'],
    'Price': [200, 1500, 80, 40, 350, 25, 300, 250, 45, 70],
    'Quantity_Sold': [1000, 500, 2000, 1500, 800, 1200, 600, 1500, 1600, 1800],
    'Rating': [4.5, 4.8, 4.0, 3.5, 4.7, 4.2, 4.3, 4.6, 4.1, 4.5]
}

df = pd.DataFrame(data)
```

```python
clothing_above_4_rating = df.loc[(df['Category'] == 'Clothing') & (df['Rating'] >= 4.0)]
average_price_clothing_above_4_rating = clothing_above_4_rating['Price'].mean()
```

```python
# 결과 출력
print(f"Clothing 카테고리에서 평점 4.0 이상인 제품의 평균 가격: {average_price_clothing_above_4_rating}")
```
