# Introduction dataset and models in sklearn

## About Iris dataset

Iris 데이터 세트는 통계학과 머신러닝 분야에서 널리 사용되는 데이터 세트 중 하나입니다. 이 데이터 세트는 1936년에 영국의 통계학자이자 생물학자인 로널드 피셔(Ronald Fisher)가 처음 소개했으며, 종종 분류(classification) 문제를 연습하는 데 사용됩니다. Iris 데이터 세트는 총 150개의 샘플로 구성되어 있으며, 이는 세 가지 종류(품종)의 Iris 꽃(세토사(Setosa), 버시컬러(Versicolor), 버지니카(Virginica)) 각각 50개의 샘플로 이루어져 있습니다.

각 샘플에는 다음과 같은 네 가지 특성이 포함되어 있습니다:

- 꽃받침 길이(Sepal Length)
- 꽃받침 너비(Sepal Width)
- 꽃잎 길이(Petal Length)
- 꽃잎 너비(Petal Width)

이 특성들은 모두 센티미터(cm) 단위로 측정됩니다. 이 데이터를 활용하여 각 Iris 꽃의 품종을 분류하는 머신러닝 모델을 훈련시킬 수 있습니다. 예를 들어, 결정 트리(decision tree), 로지스틱 회귀(logistic regression), 서포트 벡터 머신(SVM), k-최근접 이웃(k-NN)과 같은 다양한 분류 알고리즘을 사용할 수 있습니다.

Iris 데이터 세트는 데이터 과학과 머신러닝을 배우는 사람들에게 기본적인 데이터 탐색, 데이터 전처리, 모델 선택 및 모델 훈련 등의 과정을 실습할 수 있는 좋은 예제를 제공합니다. Python에서는 scikit-learn 라이브러리를 통해 이 데이터 세트를 쉽게 불러오고 사용할 수 있습니다.

# Quest with iris, digits, wine, cancer datasets

## Quest for load_iris

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris.keys()
```

```python
iris_data = iris.data
iris_label = iris.target
iris_data.shape, iris_label.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.rcParams['font.family'] = 'Avenir'



X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.3, random_state=25)

Acc = []
Prec = []
Recall = []
F1 = []

for model in [DecisionTreeClassifier(), 
              RandomForestClassifier(), 
              svm.SVC(), 
              SGDClassifier(), 
              LogisticRegression(max_iter=10000),
              XGBClassifier()
              ]:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(model.__class__.__name__, accuracy_score(y_test, y_pred))
    Acc.append(accuracy_score(y_test, y_pred))
    Prec.append(precision_score(y_test, y_pred, average='weighted'))
    Recall.append(recall_score(y_test, y_pred, average='weighted'))
    F1.append(f1_score(y_test, y_pred, average='weighted'))

fig, ax = plt.subplots(2, 1, figsize=(15, 8))

sns.barplot(y=Acc, x=['DecisionTree', 'RandomForest', 'SVM', 'SGD', 'Logistic', 'XGBClassifier'], hue = Acc, palette='crest', ax=ax[0]) 
for i, acc in enumerate(Acc):
    ax[0].text(i, acc, round(acc, 4), ha='center', va='bottom')
ax[0].legend().remove()
ax[0].set_ylim([0.6, 1])
ax[0].set_title('Accuracy of each model for Iris Dataset')

sns.lineplot(x=['DecisionTree', 'RandomForest', 'SVM', 'SGD', 'Logistic', 'XGBClassifier'], y=Prec, label='Precision', ax=ax[1], marker='o')
sns.lineplot(x=['DecisionTree', 'RandomForest', 'SVM', 'SGD', 'Logistic', 'XGBClassifier'], y=Recall, label='Recall', ax=ax[1], marker='^')
sns.lineplot(x=['DecisionTree', 'RandomForest', 'SVM', 'SGD', 'Logistic', 'XGBClassifier'], y=F1, label='F1', ax=ax[1], marker='s', markersize=10)
for i, f1 in enumerate(F1):
    ax[1].text(i, f1, round(f1, 4), ha='center', va='bottom')
ax[1].set_title('Precision/Recall/F1 of each model for Iris Dataset')
```

# Comparing model performance between toy datasets

1. Iris Dataset: Being a relatively simple and small dataset, most models perform well. However, ensemble methods like RandomForest and gradient boosting methods like XGBoost may capture relationships better due to their ability to handle the interaction of features. 
SVM might perform slightly better due to its ability to find the optimal hyperplane that separates the classes. The Iris dataset is known for having features that are linearly separable, which plays to the strengths of SVM. SVM uses a technique called kernel trick to transform the data and then based on these transformations, it finds an optimal boundary between the possible outputs. That's why, even with a small margin, SVM might outperform other models which may not find as optimal a division between the classes in a feature space.

2. Wine Dataset: The perfect scores of Decision Tree and RandomForest could be due to the dataset's features having clear and decisive boundaries for classification, which trees can easily capture.

3. Digits Dataset: High dimensionality (64 features) may be why SVM performs best, as it's effective in high-dimensional spaces, especially with clear margins of separation.

4. Cancer Dataset: Again, XGBoost shines due to its strength in handling binary classification problems and datasets where feature relationships are complex.

5. Each algorithm has its own method of finding patterns in the data, which makes them perform differently depending on the dataset's structure, size, and inherent complexities.
